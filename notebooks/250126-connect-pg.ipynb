```python
from sqlalchemy import create_engine
from sqlalchemy import text

engine = create_engine(
    "postgresql+psycopg2://root:root@localhost:5432/ny_taxi"
)

# test connection
with engine.connect() as conn:
    sql_statement = "SELECT 1"
    result = conn.execute(text(sql_statement))
    print(result.fetchone())
```

```python
import pandas as pd
import numpy as np

# Test Insert Mock data
def create_sample_mock_data(rows, cols):
    """
    Generates a sample DataFrame of random integers.
    """
    data = np.random.randint(0, 100, size=(rows, cols))
    df = pd.DataFrame(data, columns=[f'col_{i}' for i in range(cols)])
    return df

# Example usage:
sample_df = create_sample_mock_data(5, 3)
display(sample_df)
```

```python
sample_df.to_sql(
    name="sample",
    con=engine,
    if_exists='append'
)
```
